```python
df = pd.read_csv("movie_scores.csv")

###################### find na

df.isnull()  # obselete
# df.info()
df['pre_movie_score'].isna()
df[~df['pre_movie_score'].isna()]
df[df['pre_movie_score'].notna()]  # show who does not have na

df['pre_movie_score'].isna()
# for value in df.iloc[2].isna():
#     print(value)
# df.iloc[2][3] is np.NaN 

def count_null(row):
    count = 0
    for col in df.columns:
        if pd.isna(row[col]):
            count += 1
    return count
df['nan count'] = df.apply(count_null, axis=1)
display(df)
df[df['nan count'] < 6]

df[(df['pre_movie_score'].notna()) & (df['first_name'].notna())]

# easy way to count na or available
df['nan count2'] = len(df.columns) -  df.notna().sum(axis=1)
df
```

```python
df = pd.read_csv("movie_scores.csv")

###################### remove na

df.dropna()  # cut all row containing NaN

df.dropna(thresh=5)  # thresh = minimum not-NaN -- per row

df.dropna(thresh=len(df.columns) // 2)  # thresh = minimum avialable -- per row at least 50% 
df.dropna(thresh=int(len(df.columns) * 0.75))  # thresh = minimum avialable -- per row at least 75% 
# 3 3.25 4
df.dropna(thresh=len(df.columns) - 2)  # thresh = minimum avialable -- per row, no more than 2 NaN

# number of columns = number of na + number of avialable
# NaN NaN NaN NotNull

df.dropna(axis=1)  # delete all table

df.dropna(thresh=4, axis=1)  # keep where 4 or more are available

# subset
# manual filter
df1 = df[df['age'].notna()]
df1[df1['sex'].notna()]

# auto filter
df.dropna(subset=['age', 'sex', 'pre_movie_score'])  # drop if there is NaN in 1 of the columns
df.dropna(subset=['age', 'sex', 'pre_movie_score'], thresh=3)  # thresh refers to the subset, at least 3 has value -- 
#                                                                same as prev line
df.dropna(subset=['age', 'sex', 'pre_movie_score'], thresh=2)  # thresh refers to the subset, at least 2 values

df.dropna(subset=['pre_movie_score'], thresh=2)  # NOT LOGICAL -- empty
```

```python
df = pd.read_csv("movie_scores.csv")

###################### replace na

display('min', df['pre_movie_score'].min())
display('mean', df['pre_movie_score'].mean())
display('max', df['pre_movie_score'].max())
display('max value counts', df['pre_movie_score'].value_counts().max())

df['pre_movie_score'].fillna(0)
print('\nmean = ', df['pre_movie_score'].mean())
df['pre_movie_score'].fillna(df['pre_movie_score'].mean())
```
